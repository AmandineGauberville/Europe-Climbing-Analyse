## Moyenne des hauteurs des voies par pays

```python
country_mean = climbing.groupby(['Country'])['Length'].mean()
country_mean.sort_values()
```

## Premier aperçu des statistiques

```python
sns.pairplot(climbing, diag_kind = 'kde', plot_kws = {'alpha': 0.6, 's': 80, 'edgecolor': 'k'})
```

On voit une corrélation entre les plateformes et la hauteur des voies

```python
#Corrélation positive forte :

climbing[['Pitches','Length']].corr()
```

```python
#Régression linéaire :

plt.figure(figsize=(10, 10))
sns.regplot(climbing.Length, climbing.Pitches, scatter=True)
```

Une corrélation positive entre les hauteurs des voies et le nombre d'emplacement pour s'arrêter.
Hétéroscédasticité.

```python
#Voir l'ensemble des taux de corrélation

plt.figure(figsize=(20, 20))
p = sns.heatmap(climbing.corr(), annot=True, square=True)
```
